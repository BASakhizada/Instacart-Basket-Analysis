## Creating loyalty flag using transform() and loc()

```python
# we split the data into groups based on the “user_id” column.
# then we apply the transform() function on the “order_number” column to generate the maximum orders for each user.
# and we create a new column, “max_order,” into which you’ll place the results of your aggregation.
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id']) ['order_number'].transform(np.max)
```

```python
# we check whether the required changes are applied to out dataframe or not
df_ords_prods_merge.head(100)
```

## Deriving Columns with Loc()

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# in the above syntax, we assigned loyalty levels to customers based on their numer of orders. 
# customers with orders more than 40 are assigned as loyal.
# customers with orders between 10 to 40 are assigned as regular customars.
# customers with less than 10 orders are assigned as new customers. 
```

## Creating a spending flag based on average price 

The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# creating a new average_price column with average of price based on user_id
df_ords_prods_merge['average_price'] = df_ords_prods_merge.groupby(['user_id']) ['prices'].transform(np.mean)
```

```python
# checking if the changes are applied.
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] >= 10, 'loyalty_flag2'] = 'Higher Spender'
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] < 10, 'loyalty_flag2'] = 'Lower Spender'
```

```python
# checking the changes in the three targeted columns 
df_ords_prods_merge[['user_id', 'average_price', 'loyalty_flag2']].head(100)
```

```python
df_ords_prods_merge['loyalty_flag2'].value_counts(dropna = False)
```

## Creating order frequency flag 

In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
df_ords_prods_merge['median_prior_order'] = df_ords_prods_merge.groupby(['user_id']) ['days_since_prior_order'].transform(np.median)
```

```python
# checking whether the changes are applied
df_ords_prods_merge.head()
```

```python
# applying the conditions to categorize customers into three distinct groups based on frequency of their orders.
df_ords_prods_merge.loc[df_ords_prods_merge['median_prior_order'] > 20, 'loyalty_flag3'] = 'Non-frequent customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['median_prior_order'] <= 20) & (df_ords_prods_merge['median_prior_order'] > 10), 'loyalty_flag3'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['median_prior_order'] <= 10, 'loyalty_flag3'] = 'Frequent customer'
```

```python
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'loyalty_flag3']].head(100)
```

```python
# checking number of customers falling under each category
df_ords_prods_merge['loyalty_flag3'].value_counts(dropna = False)
```
