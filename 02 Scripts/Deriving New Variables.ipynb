## If-Statements with User-Defined Functions

```python
# Define a Function 
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the Function 
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# counting the frequency the newly created column
df['price_range'].value_counts(dropna = False)
```

## If-Statements with the loc() Function

### creating conditions for the price_range column

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Checking the frequency of the newly created column 
df['price_range_loc'].value_counts(dropna = False)
```

```python
# If statements with loc() runs very mich faster compared to user-defined function since it looks for the matches only in the targeted column. 
# User-defind function looks for the matches in the entire dataframe.
```

## If-Statements with For-Loops

```python
# cheking which days of the week are the busiest by checking the frequency of order_dow column. 
df_ords_prods_merge['order_dow'].value_counts(dropna = False)
```

```python
# 0 = Saturday and it's the busiest day of the week.
# 4 = Wednesday and it's the least busy day of the week.
# rest of days of the week are regularly busy.
```

```python
# Creating the loop and assigning conditions
result = []
for value in df_ords_prods_merge['order_dow']: 
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
# merging the loop into the dataframe by creating a new column titled busiest_day.
df_ords_prods_merge['busiest_day'] = result 
```

```python
# checking frequency of the busiest_day column to identify how many days are busiest, regularly busy and leas busy.
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

## Creating new columns as busiest_days and slowest_days

```python
df_ords_prods_merge['order_dow'].value_counts(dropna = False)
```

```python
# Saturday (0) and Sunday (1)are the two busiest_days of the week 
# Tuesday (3) and Wednesday (4) are the two slowest_days of the week
# Friday, Monday and Thursday are regluarly_busy_days of the week
```

```python
# Creating the loop and assigning conditions
result = []
for value in df_ords_prods_merge['order_dow']: 
    if value == 0 or value == 1:
        result.append('busiest_days')
    elif value == 3 or value == 4:
        result.append('slowest_days')
    else: 
        result.append('regularly_busy_days')
```

```python
df_ords_prods_merge['busiest_days'] = result 
```

```python
# cheking how the new column looks in the dataframe
df_ords_prods_merge.head()
```

```python
# checking count of the column to make sure the numbers are correct.
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

## Creating a new column for busiest hours of the day

```python
# locating name of the column where order hours are recoreded
df_ords_prods_merge.head()
```

```python
# Checking what hours of the day are the busiest and vice versa. 
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# dividing hours of the day based on orders received into categories
# Most orders = 10, 11, 14, 15, 13, 12, 16, 9, 17
# Average Orders = 
# Fewest Orders - 2, 4, 3, 5, 23, 6, 0, 1
```

```python
# Creating if-loop statement

result_hour = []

for value in df_ords_prods_merge['order_hour_of_day']:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_hour.append('Most orders')
  elif value in [2, 4, 3, 5, 23, 6, 0, 1]:
    result_hour.append('Fewest orders')
  else:
    result_hour.append('Average orders')
```

```python
df_ords_prods_merge['busiest_hours'] = result_hour
```
